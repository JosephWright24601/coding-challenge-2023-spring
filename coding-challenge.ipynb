```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv")
```

```python
df.head()
```

```python
#Star color and temperature 
plt.figure(figsize=(26,4), dpi=200)
sns.barplot(data=df,x='Star color', y="Temperature (K)", estimator=np.mean,ci='sd')
#plt.legend(loc=(1.05,0))
plt.show()
```

```python
#No correlation between increasing luminosity and radius
x = df['Radius(R/Ro)']
y = df['Luminosity(L/Lo)']
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel("Luminosity(L/Lo)")
axes.set_xlabel("Radius(R/Ro)")
axes.plot(x,y)
plt.show()
```

```python
#Some correlation in radius and star type
x = df['Radius(R/Ro)']
y = df['Star type']
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel("Star type")
axes.set_xlabel("Radius(R/Ro)")
axes.plot(x,y)
plt.show()
```
